## Como a renda per capita influencia no acesso à internet na região Sudeste brasileira?

___
## Introdução

        O objetivo da análise é avaliar a influência da renda per capita nos meios de acesso à internet na região Sudeste do Brasil. 
        A análise utilizará dados como a posse de meios de acesso, como telefone celular, microcomputador e tablet, e os tipos de provedores de conexão, como 3G/4G, discada por linha telefônica e banda larga, associando-os à renda per capita dos entrevistados para alcançar os resultados desejados.


___
## Análise e Resultados

```python
import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
pd.options.mode.chained_assignment = None
from numpy import zeros_like
```

```python
#Ler o dataset
dados1 = pd.read_csv("DOM2013.csv")

#separar dados de interesse
dados_resumidos1=dados1.loc[:,["UF","V4621","V0220","V02324","V0231","V02323","V02321","V02325","V02322","V02424","V02425","V02426"]]

#tirar nulos
dados_NN1=dados_resumidos1[(pd.notnull(dados_resumidos1.UF))&(pd.notnull(dados_resumidos1.V4621))&(pd.notnull(dados_resumidos1.V0220))&(pd.notnull(dados_resumidos1.V02324))&(pd.notnull(dados_resumidos1.V0231))&(pd.notnull(dados_resumidos1.V02323))&(pd.notnull(dados_resumidos1.V02321))&(pd.notnull(dados_resumidos1.V02325))&(pd.notnull(dados_resumidos1.V02322))&(pd.notnull(dados_resumidos1.V02424))&(pd.notnull(dados_resumidos1.V02425))&(pd.notnull(dados_resumidos1.V02426))]

#filtro por Unidade Federativa
dados_R1=dados_NN1[(dados_NN1.UF== 31)|(dados_NN1.UF== 32)|(dados_NN1.UF== 33)|(dados_NN1.UF== 35)]

#definir tipo de coluna
dados_R1.UF=dados_R1.UF.astype('category')
dados_R1.V4621=dados_R1.V4621.astype('float')
dados_R1.V0220=dados_R1.V0220.astype('category')
dados_R1.V02324=dados_R1.V02324.astype('category')
dados_R1.V0231=dados_R1.V0231.astype('category')
dados_R1.V02323=dados_R1.V02323.astype('category')
dados_R1.V02321=dados_R1.V02321.astype('category')
dados_R1.V02325=dados_R1.V02325.astype('category')
dados_R1.V02322=dados_R1.V02322.astype('category')
dados_R1.V02424=dados_R1.V02424.astype('category')
dados_R1.V02425=dados_R1.V02425.astype('category')
dados_R1.V02426=dados_R1.V02426.astype('category')

#substituição de labels
dados_R1.UF.cat.categories = (['MG', 'ES','RJ','SP'])
dados_R1.V0220.cat.categories = (['Sim','Não'])
dados_R1.V02324.cat.categories = (['Sim','Não'])
dados_R1.V0231.cat.categories = (['Sim','Não'])
dados_R1.V02323.cat.categories = (['Sim','Não'])
dados_R1.V02321.cat.categories = (['Sim','Não'])
dados_R1.V02325.cat.categories = (['Sim','Não'])
dados_R1.V02322.cat.categories = (['Sim'])
dados_R1.V02424.cat.categories = (['Sim','Não'])
dados_R1.V02425.cat.categories = (['Sim','Não'])
dados_R1.V02426.cat.categories = (['Sim','Não'])

#organizar unidade dos dados do tipo float
dados_R1 = dados_R1[dados_R1.V4621!=999999999999]

#mudança de nome das colunas 
dados_R1.columns = ('Unid. da Federação','Renda per capita','Tem telefone móvel celular','Para acessar à Internet utiliza-se telefone celular','Tem microcomputador','Para acessar à Internet utiliza-se microcomputador','Tem tablet','Para acessar à Internet utiliza-se tablet','Tem acesso à Internet no próprio domicilio','O acesso à Internet é feito através de sinal de rede celular 3G ou 4G','O acesso à Internet é feito através de conexão discada por linha telefônica','O acesso à Internet é feito por banda larga')

#dataset final
dados_R1.to_csv("resumo_2013.csv")

#definindo faixas de renda
baixa_renda1 = dados_R1[dados_R1["Renda per capita"]<=900]
media_renda1 = dados_R1[(dados_R1["Renda per capita"]>900)&(dados_R1["Renda per capita"]<1500)]
alta_renda1 = dados_R1[dados_R1["Renda per capita"]>=1500]

dados_R1
```

```python
faixa = range(10,3000,100)
plot2013 = dados_R1["Renda per capita"].plot.hist(bins=faixa,title='Renda per capita / 2013', weights=zeros_like(dados_R1["Renda per capita"])+1./dados_R1["Renda per capita"].size*100, figsize=(6, 6), alpha=0.7)
plt.ylabel('Frequência relativa (em %)')
plt.xlabel('renda (em reais)')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

#Para faixa de baixa renda per capita
cel_baixa1 = baixa_renda1["O acesso à Internet é feito através de sinal de rede celular 3G ou 4G"].value_counts()
cel_baixa1.plot(kind='pie', ax=axes[0,0], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[0,0].set_title('Acesso 3G/4G - Baixa renda / 2013')
disc_baixa1 = baixa_renda1["O acesso à Internet é feito através de conexão discada por linha telefônica"].value_counts()
disc_baixa1.plot(kind='pie', ax=axes[0,1], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[0,1].set_title('Acesso discada - Baixa renda / 2013')
larga_baixa1 = baixa_renda1["O acesso à Internet é feito por banda larga"].value_counts()
larga_baixa1.plot(kind='pie', ax=axes[0,2], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[0,2].set_title('Acesso Banda Larga - Baixa renda / 2013')

#Para faixa de renda média per capita
cel_media1 = media_renda1["O acesso à Internet é feito através de sinal de rede celular 3G ou 4G"].value_counts()
cel_media1.plot(kind='pie', ax=axes[1,0], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[1,0].set_title('Acesso 3G/4G - Renda média / 2013')
disc_media1 = media_renda1["O acesso à Internet é feito através de conexão discada por linha telefônica"].value_counts()
disc_media1.plot(kind='pie', ax=axes[1,1], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[1,1].set_title('Acesso discada - Renda média / 2013')
larga_media1 = media_renda1["O acesso à Internet é feito por banda larga"].value_counts()
larga_media1.plot(kind='pie', ax=axes[1,2], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[1,2].set_title('Acesso Banda Larga - Renda média / 2013')

#Para faixa de alta renda per capita
cel_alta1 = alta_renda1["O acesso à Internet é feito através de sinal de rede celular 3G ou 4G"].value_counts()
cel_alta1.plot(kind='pie', ax=axes[2,0], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[2,0].set_title('Acesso 3G/4G - Alta renda / 2013')
disc_alta1 = alta_renda1["O acesso à Internet é feito através de conexão discada por linha telefônica"].value_counts()
disc_alta1.plot(kind='pie', ax=axes[2,1], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[2,1].set_title('Acesso discada - Alta renda / 2013')
larga_alta1 = alta_renda1["O acesso à Internet é feito por banda larga"].value_counts()
larga_alta1.plot(kind='pie', ax=axes[2,2], figsize=(30,30), colors=('blue', 'red'), autopct="%0.2f",legend=False);axes[2,2].set_title('Acesso Banda Larga - Alta renda / 2013')
```

Descreva aqui suas conclusões!

    De início, é importante ressaltar que a média da renda per capita da população, nos três anos avaliados, se manteve praticamente igual, perto dos 750 reais para boa parte da população.
    O que pode ser observado com os dados obtidos é que o uso de internet por meio de conexão 3G/4G sofreu um grande aumento, independente da faixa salarial, cerca de 21% de aumento no uso deste tipo de conexão de 2013 para 2015, aumento dados provavelmente pela praticidade do serviço e o custo relativamente baixo deste.
    Para a conexão do tipo discada por linha telefônica, a média de uso da população manteve-se próxima, cerca de 96% da população tem acesso a esse tipo de conexão, independente da faixa salarial.
    A conexão banda larga foi a que apresentou uma variação mais palpável de análise. Para a faixa salarial de baixa renda, até 900 reais, houve uma queda de cerca de 10%, que pode ser explicada, provavelmente, por um aumento no custo de tal serviço, ou a diminuição do poder de consumo, o que levou ao cancelamento do serviço para possível redução de custo. Para a faixa de renda média, a queda foi de 8%, provavelmente pelos mesmos motivos da baixa renda, e para a alta renda, a variação foi baixa, cerca de 3%.
    Desta forma, é possível concluir que a renda per capita tem certa influência nos meios de acesso e serviços provedores de internet, porém, dado o fato de que a média da renda sofreu baixa alteração, o que pode ser explicado foi o possível aumento da inflação e diminuição de consumo, levando os consumidores a abandonarem tais serviços para reduzir custos e destiná-los a outros investimentos.
