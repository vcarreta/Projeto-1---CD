# Projeto 1 - Ciência dos dados 

### O tema escolhido foi acesso à Internet e à Televisão. O objetivo é detectar mudanças na acessibilidade à internet com base nos meios de uso e nos tipos de acesso.

```python
import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
pd.options.mode.chained_assignment = None

#Ler o dataset
dados = pd.read_csv("DOM2015.csv")

#separar dados de interesse
dados_resumidos=dados.loc[:,["UF","V4621","V0220","V02324","V0231","V02323","V02321","V02325","V02322","V02424","V02425","V02426"]]

#tirar nulos
dados_NN=dados_resumidos[(pd.notnull(dados_resumidos.UF))&(pd.notnull(dados_resumidos.V4621))&(pd.notnull(dados_resumidos.V0220))&(pd.notnull(dados_resumidos.V02324))&(pd.notnull(dados_resumidos.V0231))&(pd.notnull(dados_resumidos.V02323))&(pd.notnull(dados_resumidos.V02321))&(pd.notnull(dados_resumidos.V02325))&(pd.notnull(dados_resumidos.V02322))&(pd.notnull(dados_resumidos.V02424))&(pd.notnull(dados_resumidos.V02425))&(pd.notnull(dados_resumidos.V02426))]

#filtro por Unidade Federativa
dados_R=dados_NN[(dados_NN.UF== 31)|(dados_NN.UF== 32)|(dados_NN.UF== 33)|(dados_NN.UF== 35)]

#definir tipo de coluna
dados_R.UF=dados_R.UF.astype('category')
dados_R.V4621=dados_R.V4621.astype('float')
dados_R.V0220=dados_R.V0220.astype('category')
dados_R.V02324=dados_R.V02324.astype('category')
dados_R.V0231=dados_R.V0231.astype('category')
dados_R.V02323=dados_R.V02323.astype('category')
dados_R.V02321=dados_R.V02321.astype('category')
dados_R.V02325=dados_R.V02325.astype('category')
dados_R.V02322=dados_R.V02322.astype('category')
dados_R.V02424=dados_R.V02424.astype('category')
dados_R.V02425=dados_R.V02425.astype('category')
dados_R.V02426=dados_R.V02426.astype('category')

#substituição de labels
dados_R.UF.cat.categories = (['MG', 'ES','RJ','SP'])
dados_R.V0220.cat.categories = (['Sim','Não'])
dados_R.V02324.cat.categories = (['Sim','Não'])
dados_R.V0231.cat.categories = (['Sim','Não'])
dados_R.V02323.cat.categories = (['Sim','Não'])
dados_R.V02321.cat.categories = (['Sim','Não'])
dados_R.V02325.cat.categories = (['Sim','Não'])
dados_R.V02322.cat.categories = (['Sim'])
dados_R.V02424.cat.categories = (['Sim','Não'])
dados_R.V02425.cat.categories = (['Sim','Não'])
dados_R.V02426.cat.categories = (['Sim','Não'])

#organizar unidade dos dados do tipo float
dados_R = dados_R[dados_R.V4621!=999999999999]

#mudança de nome das colunas 
dados_R.columns = ('Unid. da Federação','Renda per capita','Tem telefone móvel celular','Para acessar à Internet utiliza-se telefone celular','Tem microcomputador','Para acessar à Internet utiliza-se microcomputador','Tem tablet','Para acessar à Internet utiliza-se tablet','Tem acesso à Internet no próprio domicilio','O acesso à Internet é feito através de sinal de rede celular 3G ou 4G','O acesso à Internet é feito através de conexão discada por linha telefônica','O acesso à Internet é feito por banda larga')

dados_R
```

```python
gp1 = dados_R["Tem acesso à Internet no próprio domicilio"].value_counts()
gp1.plot(kind='pie', colors=('blue', 'red'), autopct="%0.2f",legend=False)
```

```python
gb1 = dados_R["O acesso à Internet é feito através de sinal de rede celular 3G ou 4G"].value_counts()
gb1.plot(kind='bar', color=('blue', 'red'), legend=False)
```

```python
gb2 = dados_R["O acesso à Internet é feito através de conexão discada por linha telefônica"].value_counts()
gb2.plot(kind='bar', color=('blue', 'yellow'), legend=False)
```

```python
gb3 = dados_R["O acesso à Internet é feito por banda larga"].value_counts()
gb3.plot(kind='bar', color=('green', 'red'), legend=False)
```

```python
ct1 = pd.crosstab(dados_R["Tem telefone móvel celular"], dados_R["Para acessar à Internet utiliza-se telefone celular"])
ct1.plot(kind='bar', color=('yellow', 'green'))
plt.ylabel('Tem celular e o utiliza para acessar a internet')
```

```python
ct2 = pd.crosstab(dados_R["Para acessar à Internet utiliza-se microcomputador"], dados_R["Tem microcomputador"])
ct2.plot(kind='bar', color=('yellow', 'green'))
plt.ylabel('Tem computador e o utiliza para acessar a internet')
```
